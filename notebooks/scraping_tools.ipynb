```python
from bs4 import BeautifulSoup
import json
import requests
```

```python
html_doc = requests.get('https://www.foxsports.com/stories/college-basketball/top-25-players-in-college-basketball-through-the-first-25-days-of-the-2023-24-season')
```

```python
html_doc= html_doc.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
scripts = soup.find_all('script', {"type" : "application/ld+json"}) 
print(scripts)
for script in scripts:
    data = json.loads(script.string)  
    article_body = data['articleBody']
    print(article_body)  # Prints the article body
```
