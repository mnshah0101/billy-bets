```python
from langchain.tools import Tool
from langchain.agents import AgentExecutor, create_structured_chat_agent
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain import hub

```

```python
def objective(query):
    print("hello there, this is objective")
    print(query)
def subjecitive(query):
    print("hello there, this is subjective")
    print(query)
subjective = Tool(
    name="Subjective Sports Questions",
    description="Useful for answering subjective based sports questions, such as expert opinion or anything that can't be answered with straigth facts.",
    func=subjecitive,
)
objective = Tool(
    name="Objective Sports Questions",
    description="Useful for answering objective based sports questions, such as finding a paticular statistic",
    func=subjecitive,
)
tools = [subjective, objective]
llm = ChatOpenAI(
    model_name='gpt-3.5-turbo',
    temperature=0.0
)
```

```python
prompt = {
    "input_variables": [
        "question"
    ],
    "output_parser": None,
    "template": "You are a conversational AI expert sports betting assistant named Billy that answers user's sports questions. Here is the user's question:\n{question}",
    "template_format": "f-string"
}
```

```python
agent = create_structured_chat_agent(llm, tools, prompt=prompt)

```

```python
agent_executor = AgentExecutor(
    agent=agent, tools=tools, verbose=True, handle_parsing_errors=True
)
```

```python
agent.invoke("What are experts saying about Messi's chance to win the ballon dor")
```
